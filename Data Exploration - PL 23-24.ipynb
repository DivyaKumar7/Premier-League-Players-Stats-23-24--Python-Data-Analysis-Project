```python
pd.options.display.max_columns = None
pd.options.display.max_rows = None

df = pd.read_csv("premier-player-23-24.csv")
df.sample(5)
```

# Dataset Description

This dataset contains detailed data on all footballers from the 2023/24 Premier League season.

## Columns

- **Player**: The name of the player.
- **Nation**: The player's nationality.
- **Pos**: The player's position (e.g., forward, midfielder, defender).
- **Age**: The player's age.
- **MP (Minutes Played)**: Total minutes played by the player.
- **Starts**: Number of matches the player started.
- **Min (Minutes)**: Total minutes played by the player (this might be the same as MP).
- **90s (90s Played)**: The equivalent of 90-minute matches played by the player (e.g., 1.5 = 135 minutes).
- **Gls (Goals)**: Total number of goals scored by the player.
- **Ast (Assists)**: Total number of assists made by the player.
- **G+A (Goals + Assists)**: Total number of goals and assists combined.
- **G-PK (Goals - Penalty Kicks)**: Total number of goals scored excluding penalty kicks.
- **PK (Penalty Kicks)**: Number of penalty goals scored by the player.
- **PKatt (Penalty Kicks Attempted)**: Number of penalty kicks attempted by the player.
- **CrdY (Yellow Cards)**: Number of yellow cards received by the player.
- **CrdR (Red Cards)**: Number of red cards received by the player.
- **xG (Expected Goals)**: The expected number of goals from the player's shots.
- **npxG (Non-Penalty Expected Goals)**: Expected goals excluding penalties.
- **xAG (Expected Assists)**: The expected number of assists from the player's passes.
- **npxG+xAG (Non-Penalty xG + xAG)**: Total of non-penalty expected goals and expected assists.
- **PrgC (Progressive Carries)**: Number of times the player carried the ball forward.
- **PrgP (Progressive Passes)**: Number of passes made by the player that moved the ball forward.
- **PrgR (Progressive Runs)**: Number of times the player made runs forward with the ball.
- **Gls (Goals)**: *(Repeated, already defined)* Total number of goals scored.
- **Ast (Assists)**: *(Repeated, already defined)* Total number of assists made.
- **G+A (Goals + Assists)**: *(Repeated, already defined)* Total number of goals and assists combined.
- **G-PK (Goals - Penalty Kicks)**: *(Repeated, already defined)* Goals scored excluding penalty kicks.
- **G+A-PK (Goals + Assists - Penalty Kicks)**: Total goals and assists minus penalty goals.
- **xG (Expected Goals)**: *(Repeated, already defined)* Expected number of goals from the player's shots.
- **xAG (Expected Assists)**: *(Repeated, already defined)* Expected number of assists from the player's passes.
- **xG+xAG (Expected Goals + Expected Assists)**: Total expected goals and assists.
- **npxG (Non-Penalty Expected Goals)**: *(Repeated, already defined)* Expected goals excluding penalties.
- **npxG+xAG (Non-Penalty xG + Expected Assists)**: Total of non-penalty expected goals and expected assists.

```python
# Lets find the most efficient top 30 attackers based on Goals and Assists per 90.

top30_attackers = df[df['MP']>=20][['Player', 'Team', 'Pos', 'G+A_90', 'xG+xAG_90', 'Gls_90', 'Ast_90', 'MP']]\
                    .sort_values('G+A_90', ascending=False).head(30)
top30_attackers
```

```python
# Lets find the most balanced top 10 attackers based on Goals and Assists per 90.

top30_attackers = top30_attackers.head(10)

plt.figure(figsize=(14, 5))
sns.scatterplot(data=top30_attackers, x='Gls_90', y='Ast_90', s=100, color='blue')

for i in range(len(top30_attackers)):
    plt.annotate(list(top30_attackers['Player'])[i],
                 (list(top30_attackers['Gls_90'])[i], list(top30_attackers['Ast_90'])[i]),
                 textcoords="offset points",
                 xytext=(10, 15),  # Offset label position
                 ha='center',  # Horizontal alignment
                 fontsize=12,
                 color='black',
                 rotation=15)

plt.xlabel("Goals per 90", fontsize=12)
plt.ylabel("Assists per 90", fontsize=12)

plt.show()
```

```python
# Lets find the most balanced top 10 attackers based on Goals and Assists.

plt.figure(figsize=(14, 5))
sns.scatterplot(data=top_attackers, x='Gls', y='Ast', s=100, color='blue')

for i in range(len(top_attackers)):
    plt.annotate(list(top_attackers['Player'])[i],
                 (list(top_attackers['Gls'])[i], list(top_attackers['Ast'])[i]),
                 textcoords="offset points",
                 xytext=(5, -35),  # Offset label position
                 ha='center',  # Horizontal alignment
                 fontsize=12,
                 color='black',
                rotation=345)

plt.xlabel("Goals", fontsize=12)
plt.ylabel("Assists", fontsize=12)

plt.show()
```

Cole Palmer has been very efficient and also well-balanced among all the top players. Though, Erling Haaland is the top scorer, we can consider Cole Palmer as the best attacker of the Premier League 23-24 season

```python
# Non-Penalty Goals

top_attackers = df[['Player', 'Team', 'Pos', 'Gls', 'G-PK', 'PK', 'npxG', 'G+A']].sort_values('G+A', ascending=False).head(15)
top_attackers
```

```python
bar_width = 0.4
x = np.arange(len(top_attackers['Player']))  # Positions for bars

plt.figure(figsize=(10, 5))

# Plot bars
plt.bar(x, top_attackers['Gls'], width=bar_width, label='Goals', color='skyblue')
plt.bar(x, top_attackers['PK'], width=bar_width, label='Penalty Goals', color='darkblue')

# Add labels and title
plt.xticks(x, top_attackers['Player'], rotation=90, fontsize=10)  # Add player names to x-axis
plt.title("Bar Chart: Goals and Penalty Goals", fontsize=16)
plt.xlabel("Players", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.legend(title="Metrics", fontsize=10)

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Top attacking Teams

Team_df = df.groupby('Team').agg({'G+A': 'sum', 'xG+xAG': 'sum'}).sort_values('G+A',ascending=False)
Team_df['xG_diff'] = Team_df['G+A']-Team_df['xG+xAG']
Team_df['xG_diff_percentage'] = Team_df['xG_diff']*100/Team_df['xG+xAG']
Team_df
```

```python
# Lets visualize the most overperforming and under performing teams

Team_df = Team_df.sort_values('xG_diff_percentage', ascending=False)

plt.figure(figsize=(10, 5))
sns.barplot(y='Team', x='xG_diff_percentage', data=Team_df, color='purple')
plt.xlabel('Overperformance/Underperformance (in %)')
plt.show()
```

### Most Overperforming Teams:
1. Manchester City
2. Crystal Palace
3. West Ham United

Manchester City overperforming the most is one of the most important factors that led them to their league title.

```python
# Lets visualize the most overperforming and under performing Goal scorers from the Top 30

Top30_attackers = df.sort_values('G+A', ascending=False, ignore_index=True).head(30)[['Player', 'Gls', 'xG','npxG', 'G+A']]
Top30_attackers['xG_diff_percentage'] = (Top30_attackers['Gls']-Top30_attackers['xG'])*100/Top30_attackers['xG']
Top30_attackers.sort_values('xG_diff_percentage', ascending=False, inplace=True)
Top30_attackers
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(y='Player', x='xG_diff_percentage', data=Top30_attackers, color='purple')
plt.xlabel('Overperformance/Underperformance (in %)')
plt.show()
```

### Observation:
Rodri, Declan Rice & Phil Foden has the overperformed the most.

```python
# Lets analyze the progressive carries and progressive runs

most_progressive_players = df[df['MP']>=20][['Player', 'PrgC', 'PrgP']]
most_progressive_players['PrgA'] = most_progressive_players['PrgC'] + most_progressive_players['PrgP']
most_progressive_players = most_progressive_players.sort_values('PrgA', ascending=False).head(30)
most_progressive_players
```

```python
plt.figure(figsize=(16, 10))
sns.scatterplot(x='PrgC', y='PrgP', data=most_progressive_players)

for i in range(len(most_progressive_players)):
    plt.annotate(list(most_progressive_players['Player'])[i],
                 (list(most_progressive_players['PrgC'])[i], list(most_progressive_players['PrgP'])[i]),
                 textcoords="offset points",
                 xytext=(5,5),  # Offset label position
                 ha='center',  # Horizontal alignment
                 fontsize=12,
                 color='black',
                rotation=15)

plt.xlabel('Progressive Carries')
plt.ylabel('Progressive Passes')
plt.show()
```

### Observation:
- Rodri has the most progressive passes.
- Jeremy Doku has the most progressive carries.
- Cole Palmer is well-balanced here as well.
